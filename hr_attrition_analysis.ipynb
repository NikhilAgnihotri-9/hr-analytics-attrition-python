## Day 2: Data Loading and Initial Inspection

- Dataset loaded successfully
- Reviewed column names, data types, and missing values
- No data cleaning performed yet


```python
selected_columns = [
    "Age",
    "Department",
    "JobRole",
    "MonthlyIncome",
    "YearsAtCompany",
    "Attrition"
]

df_clean = df[selected_columns].copy()
```

```python
df_clean.info()
```

```python
df_clean.isnull().sum()
```

## No Missing Values Were Found.

```python
df_clean.isnull().sum()

```

```python
df_clean["AttritionFlag"] = df_clean["Attrition"].map({
    "Yes": 1,
    "No": 0
})
```

```python
df_clean[["Attrition", "AttritionFlag"]].head()
```

## Day 3: Data Cleaning and Column Selection

- Selected core columns relevant to HR attrition analysis
- Created a clean working dataset
- Verified data types and handled missing values
- Converted attrition into a numeric flag for analysis


```python
dept_employee_count = (
    df_clean
    .groupby("Department")
    .size()
    .reset_index(name="EmployeeCount")
)

dept_employee_count
```

```python
dept_attrition_rate = (
    df_clean
    .groupby("Department")["AttritionFlag"]
    .mean()
    .reset_index(name="AttritionRate")
)

dept_attrition_rate
```

```python
dept_summary = dept_employee_count.merge(
    dept_attrition_rate,
    on="Department"
)

dept_summary
```

```python
dept_summary["AttritionRatePercent"] = (
    dept_summary["AttritionRate"] * 100
).round(2)

dept_summary
```

## Day 4: Department-wise Employee Count and Attrition

Key observations:
- The department with the highest employee count is Sales.
- The department with the highest attrition rate is Sales.
- Departments with lower headcount but higher attrition may require targeted retention strategies is Human Resources.

```python
salary_attrition = (
    df_clean
    .groupby("Attrition")["MonthlyIncome"]
    .mean()
    .reset_index()
)
salary_attrition
```

```python
plt.Figure()
plt.bar(
    salary_attrition["Attrition"],
    salary_attrition["MonthlyIncome"]
)
plt.title("Average Monthly Income by Attrition Status")
plt.xlabel("Attrition")
plt.ylabel("Average Monthly Income")
plt.show()
```

```python
age_attrition = (
    df_clean
    .groupby("Attrition")["Age"]
    .mean()
    .reset_index()
)
age_attrition
```

```python
plt.figure()
plt.bar(
    age_attrition["Attrition"],
    age_attrition["Age"]
)
plt.title("Average Age by Attrition Status")
plt.xlabel("Attrition")
plt.ylabel("Average Age")
plt.show()
```

```python
tenure_attrition = (
    df_clean
    .groupby("Attrition")["YearsAtCompany"]
    .mean()
    .reset_index()
)
tenure_attrition
```

```python
plt.figure()
plt.bar(
    tenure_attrition["Attrition"],
    tenure_attrition["YearsAtCompany"]
)
plt.title("Average Tenure by Attrition Status")
plt.xlabel("Attrition")
plt.ylabel("Years at Company")
plt.show()
```

## Day 5: Attrition vs Salary, Age, and Tenure

Key observations:
- Employees who left tend to have lower/higher average monthly income compared to those who stayed.
- Attrition appears higher among younger/older employees.
- Employees with shorter/longer tenure are more likely to leave.

These patterns suggest that compensation structure, career stage, and employee tenure play a role in attrition risk.


```python
role_attrition = (
    df_clean
    .groupby("JobRole")["AttritionFlag"]
    .mean()
    .reset_index(name="AttritionRate")
)
role_attrition
```

```python
role_attrition["AttritionRatePercent"] = (
    role_attrition["AttritionRate"] * 100
).round(2)

role_attrition
```

```python
plt.figure()
plt.bar(
    role_attrition["JobRole"],
    role_attrition["AttritionRatePercent"]
)

plt.title("Attrition Rate by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Attrition Rate (%)")
plt.xticks(rotation=45, ha="right")

plt.show()
```

```python
dept_tenure = (
    df_clean
    .groupby("Department")["YearsAtCompany"]
    .mean()
    .reset_index(name="AverageTenure")
)

dept_tenure
```

```python
plt.figure()
plt.bar(
    dept_tenure["Department"],
    dept_tenure["AverageTenure"]
)
plt.title("Average Tenure by Department")
plt.xlabel("Department")
plt.ylabel("Years at Company")
plt.show()
```

## Day 6: Job Role Attrition Risk and Department Tenure

Key observations:
- Certain job roles exhibit significantly higher attrition rates, indicating role-specific retention challenges.
- Departments with lower average tenure may be experiencing onboarding, workload, or career progression issues.
- HR teams can prioritize retention strategies for high-risk roles and departments with shorter tenure.


## Consolidated Insights
- Departments such as Sales show higher attrition rates despite having 446 employee count.
- Employees who left tend to have lower average salary compared to those who stayed.
- Attrition is more common among employees with 5 year tenure.
- Certain job roles such as Sales Representative, Laboratory Technician, Human Resources, Sales Executive, Research Scientist exhibit significantly higher attrition risk.
- Departments like Research & Development have lower average tenure, indicating potential retention issues.


## HR Recommendations
1. Focus retention initiatives on high-attrition job roles by reviewing workload, role expectations, and growth opportunities.
2. Evaluate compensation structures for roles or departments where attrition is higher among lower-paid employees.
3. Strengthen onboarding and early-career engagement programs to reduce attrition among employees with shorter tenure.
4. Conduct department-specific engagement surveys to understand local attrition drivers.


## Executive Summary
This analysis examined employee attrition patterns across departments, job roles, compensation levels, age, and tenure. The findings indicate that attrition is not evenly distributed and is more pronounced in specific roles and early tenure stages. Compensation and career-stage factors appear to influence attrition risk. Based on these insights, targeted retention strategies focusing on high-risk roles and early employee engagement can help reduce overall attrition.
